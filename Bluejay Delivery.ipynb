```python
import pandas as pd
from openpyxl import load_workbook
from datetime import timedelta
```

```python
# Load the Excel file
file_path = '/content/assignment.xlsx'
df = pd.read_excel(file_path)
```

```python
df
```

```python
# Sort the DataFrame by Employee Name and Time
df.sort_values(by=['Employee Name', 'Time'], inplace=True)

# Convert 'Time' and 'Time Out' columns to datetime objects for time calculations
df['Time'] = pd.to_datetime(df['Time'])
df['Time Out'] = pd.to_datetime(df['Time Out'])

# Group the data by 'Employee Name'
grouped = df.groupby('Employee Name')

# Initialize variables to keep track of consecutive days and previous shift's end time
consecutive_days = 0
prev_end_time = None

# Loop through each employee's data
for name, group in grouped:
    for index, row in group.iterrows():
        # Calculate the time difference between shifts
        if prev_end_time is not None:
            time_difference = (row['Time'] - prev_end_time).total_seconds() / 3600  # in hours
            if 1 < time_difference < 10:
                print(f"{name} has less than 10 hours between shifts but greater than 1 hour.")
        # Check if the employee worked for more than 14 hours in a single shift
        shift_duration = (row['Time Out'] - row['Time']).total_seconds() / 3600  # in hours
        if shift_duration > 14:
            print(f"{name} has worked for more than 14 hours in a single shift.")
        # Check for consecutive days
        if prev_end_time is not None and (row['Time'] - prev_end_time).days == 1:
            consecutive_days += 1
            if consecutive_days == 6:
                print(f"{name} has worked for 7 consecutive days.")
        else:
            consecutive_days = 0  # Reset consecutive days count
        prev_end_time = row['Time Out']
```
